```python
import matplotlib.pyplot as plt

x = ['a', 'b', 'c', 'd', 'e']
y = [1, 14, 5, 7, 25]

# create a basic line plot

# plt.plot(x, y)
# plt.xlabel('Names')
# plt.ylabel('Values')
# plt.title('Basic Line Plot')
# plt.show()

# create a customized line plot
plt.plot(x, y, color='red', linestyle='--', marker='o', linewidth=1)
plt.grid(True)
```
