```python
import pandas as pd

df = pd.read_csv('data.csv')

# fetch the first 5 rows
df.head(5)
```

```python
# statistical analysis
df.describe()

# df.dtypes
```

```python
# handling missing values

df.isnull().any()

df.isnull().sum()

df.fillna(10.0)

# filling in missing values

df['Sales_filled'] = df['Sales'].fillna(df['Sales'].mean())
df
```

```python
df.dtypes

# renaming columns

df = df.rename(columns={'Date': 'Sales Date'})
df.head()
```

```python
# change the data types

df['Value New'] = df['Value'].fillna(df['Value'].mean()).astype(int)

df.head()
```

```python
### data aggregating and grouping

grouped_data = df.groupby('Product')['Value'].mean()
print(grouped_data)

grouped_data_mul = df.groupby(['Product', 'Region'])['Value'].sum()
print(grouped_data_mul)

grouped_data_aggregated = df.groupby('Region')['Value'].agg(['sum', 'mean'])
grouped_data_aggregated
```

```python
### merging dataframes

df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1,2,3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4,5,6]})

pd.merge(df1, df2, on='Key', how='right')
```
